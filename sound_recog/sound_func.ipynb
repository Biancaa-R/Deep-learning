Various audio functions:

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
filepath="D:\\sound_recog\\music\\asavari\\asavari02.wav"
data,sample_rate=librosa.load(filepath)
plt.figure(figsize=(12,4)) #setting the size in inches
librosa.display.waveshow(data,sr=sample_rate)
print(sample_rate) #22050 is always the sample rate
```

```python
print(data.shape)
```

```python
X = librosa.stft(data)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14, 4))
librosa.display.specshow(Xdb, sr=sample_rate, x_axis='time', y_axis='hz')
plt.colorbar()
```

```python
plt.figure(figsize=(14, 4))
plt.set_xlabel="hello" #its not displaying!
librosa.display.specshow(Xdb, sr=sample_rate, x_axis='time', y_axis='log')
plt.colorbar()
```
